```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
raw_data = os.path.join('..', 'src', 'analytics', 'rating_price.csv')
df = pd.read_csv(raw_data)
display(df)
```

```python
# grafico de relaçao linear entre faixa de rating x preço medio 
plt.figure(figsize=(15,8))
plt.title('relaçao linear entre faixa de rating x preço medio')
sns.scatterplot(data=df, x='review_scores_rating', y='preco_medio')
plt.xlabel('Faixa de Rating')
plt.ylabel('Preço medio')
plt.tight_layout()
plt.show()
```
