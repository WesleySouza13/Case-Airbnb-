```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import os 
import numpy as np 
```

```python
data_path = os.path.join('..', 'src', 'analytics', 'HostLifeCycle.csv')
df = pd.read_csv(data_path)
display(df)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
plt.title('mapa de valores nulos')
sns.heatmap(df.isna())
```

```python
some_cols = ['weekly_price', 'monthly_price', 'security_deposit', 'cleaning_fee']
df[some_cols].describe()
```

```python
df[some_cols] = df[some_cols].fillna(df[some_cols].median())
```

```python
df[some_cols].describe()
```

```python
df.describe()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
# separando tipos de dados 
num_col = df.select_dtypes(exclude='object')
cat_col = df.select_dtypes(exclude='number').columns
```

```python
for i in cat_col:
    uni = df[i].unique()
    print(f'valores unicos em {i}: {uni}')
```

```python
#verificando distribuiçoes das colunas numericas 
for num in num_col:
    plt.figure(figsize=(10,5))
    plt.title(f'Histograma coluna {num}')
    sns.histplot(data=df, x=num, edgecolor='black', kde=True)
    plt.xlabel(f'{num}')
    plt.ylabel('frequencia')
    plt.tight_layout()
    plt.show()
```

```python
# plotando graficos de caixa 
for box in num_col:
    plt.figure(figsize=(10,5))
    plt.title(f'Boxplot coluna {box}')
    sns.boxplot(data=df, y=box)
    plt.xlabel(f'{box}')
    plt.tight_layout()
    plt.show()
```

```python
# contando colunas catgoricas 
for cat in cat_col:
    plt.figure(figsize=(10,5))
    plt.title(f'Distribuiçao coluna {cat}')
    sns.countplot(data=df, x=cat)
    plt.xlabel(f'cat')
    plt.ylabel('contagem')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
    
```

```python
# grafico de correlaçao 
plt.figure(figsize=(15,10))
sns.heatmap(num_col.corr(), annot=True, cmap='coolwarm', 
            vmin=-1, vmax=1, center=0,  cbar=True, fmt=".2f", square=True, linewidths=.5)
```
